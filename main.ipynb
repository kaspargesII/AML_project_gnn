```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import pickle

with open('./dataset/Processed_data/sub001.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
data.shape
```

```python
sample_video = data[0]

# Create a new figure with a specified size
fig, ax = plt.subplots(32,1, figsize = (20,20))

# Loop over each channel
for i in range(sample_video.shape[0]):
    # Plot the data for this channel
    ax[i].plot(sample_video[i])

# Set the title and labels


# Display the plot
plt.show()
```
